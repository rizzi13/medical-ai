## Cell 2 — Download the Dataset 📥
We use a real Disease-Symptom dataset. Each row = one disease + its symptoms.

```python
import urllib.request
import os

DATA_URL = 'https://raw.githubusercontent.com/itachi9604/healthcare-chatbot/master/Data/dataset.csv'
DATA_FILE = 'disease_symptoms.csv'

if not os.path.exists(DATA_FILE):
    print('📥 Downloading dataset...')
    urllib.request.urlretrieve(DATA_URL, DATA_FILE)
    print('✅ Download complete!')
else:
    print('✅ Dataset already exists locally.')

# Load into pandas
df = pd.read_csv(DATA_FILE)
print(f'\n📊 Dataset loaded: {df.shape[0]} rows, {df.shape[1]} columns')
```

## Cell 3 — Explore the Data 🔍
Let's see what the data looks like!

```python
# First 5 rows
print('🔍 First 5 rows:')
df.head()
```

```python
# What diseases are in our dataset?
print(f'Total diseases: {df["Disease"].nunique()}')
print('\nAll diseases:')
for d in sorted(df['Disease'].unique()):
    print(f'  • {d}')
```

```python
# How many records per disease?
print('Records per disease:')
df['Disease'].value_counts()
```

## Cell 4 — Visualize 📊
A bar chart of all diseases in our dataset.

```python
plt.figure(figsize=(14, 8))
disease_counts = df['Disease'].value_counts()
sns.barplot(x=disease_counts.values, y=disease_counts.index, palette='viridis')
plt.title('🩺 Number of Records per Disease', fontsize=16, fontweight='bold')
plt.xlabel('Number of Records')
plt.ylabel('Disease')
plt.tight_layout()
plt.show()
print('✅ Day 4 COMPLETE! Tomorrow: Data Cleaning & Feature Engineering')
```
